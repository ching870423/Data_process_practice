# Training/Testing data
- A method to measure the accuracy of your model.
- To split the dtat set into 2 sets: training set & testing set.
- usually 80% for training (create the model), 20% for testing (test the accuracy of the model)

```python
# import library
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score
```

```python
# create a random data set:  100 customers in a shop, and their shopping habits.
# to get the deterministic random data you want
np.random.seed(2)
# x: the number of minutes before making a purchase
x = np.random.normal(3, 1, 100)
# y: the amount of money spent on the purchase
y = np.random.normal(150, 40, 100)/x

# draw scatter plot
plt.plot(x, y, 'o', label='original data', color='#E9E444')
plt.legend()
plt.show
```

# Splt into training/testing data

```python
# Split into Train/Test
train_x = x[:80]
train_y = y[:80]

test_x = x[80:]
test_y = y[80:]
```

```python
# Display training set
plt.plot(train_x, train_y, 'o', label='training data', color='#449FE9')
plt.legend()
plt.show()
```

```python
# Display the testing data
plt.plot(test_x, test_y, 'o', label='testong data', color='#E99D44')
plt.legend()
plt.show()
```

# Fit a polynomial model

```python
# using training data to fit a polynomial regression model
polymodel = np.poly1d(np.polyfit(train_x, train_y, 4))
polyline_x = np.linspace(0, 6, 100)

plt.plot(train_x, train_y, 'o', label='training data', color='#449FE9')
plt.plot(polyline_x, polymodel(polyline_x), 'r', label='fitting line for training data', color='#15B611')
plt.legend()
plt.show()
```

# Evaluate the fitting of model

```python
# R-squared: how well does my training data fit a polynomial regression
# Compared between real y values & calculated y values
r_squared = r2_score(train_y, polymodel(train_x))
print("R-squared of the polynomial model=", r_squared)
```

```python
# To see how well does my testing data fit a polynomial model
r_squared_test = r2_score(test_y, polymodel(test_x))
print("R-squared of the polynomial model for testing data=", r_squared_test)
```

# Predict values

```python
# How much money will a buying customer spend, if she or he stays in the shop for 5 minutes?
print("Expected spend=", polymodel(5))
```

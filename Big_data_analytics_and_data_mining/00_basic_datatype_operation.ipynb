# String
reference source: https://www.w3schools.com/python/python_strings.asp

### Indexing

source: https://medium.com/python4u/hello-python-509eabe5f5b1

```python
# Sllcing
temp = "  Hello, World!"
print(temp)
# get the characters from position 2 to position 5(not include)
print(temp[2:5])

# Negative indexing
print(temp[-5:-2])

# String length
print("String length:", len(temp))

# Removes any whitespace only from the beginning or the end
print(temp.strip())

# lower case
print("lower case:", temp.lower())

# upper case
print("upper case:", temp.upper())

# replaces a string with another string
print(temp.replace("H","J"))

# splits the string into substrings if it finds instances of the separator
print(temp.split(","))

# Check String
txt = "The rain in Spain stays mainly in the plain."
in_txt = "ain" in txt
if in_txt:
    print("\"ain\" is in txt.")
else:
    print("\"ain\" is not in txt.")

not_in_txt = "ain" not in txt
if not_in_txt:
    print("\"ain\" is not in txt.")
else:
    print("\"ain\" is in txt.")
    
# be placed into the respective placeholders
quantity = 3
itemno = 567
price = 49.95
myorder = "I want to pay {2} dollars for {0} pieces of item {1}."
print(myorder.format(quantity, itemno, price))
```

# List
reference source: https://www.w3schools.com/python/python_lists.asp

```python
# create a list
fruit = ["apple", "banana", "cherry", "orange", "kiwi","melon", "mango"]
print(fruit)

# print the second item
print(fruit[1])
print(fruit[2:5])
print(fruit[:4])
print(fruit[2:])

# loop through a list
for x in fruit:
    print(x)
    
# check if "apple" is present in the list
if "apple" in fruit:
    print("Yes, 'apple' is in the fruits list.")
    
# how many items in a list
print("list length:", len(fruit))

# add items
fruit.append("peach")
print("Add peach:", fruit)
fruit.insert(1, "blueberry")
print("Add blueberry:", fruit)

# remove item
fruit.remove("banana")
print("No banana:", fruit)
fruit.pop(1)
print("No the second fruit:", fruit)
del fruit[0]
print("No the first fruit:", fruit)

# empty the list
fruit.clear()
print("No any fruit:", fruit)

# copy a list
fruit = ["apple", "banana", "cherry"]
fruit_copy = fruit.copy()
print(fruit_copy)
fruit_list = list(fruit)
print(fruit_list)

# join 2 lists
list_alp = ["a", "b", "c"]
list_num = [1, 2, 3]
list_alp.extend(list_num)
print(list_alp)

# sort the list
list_num = [4, 23, 1, 3, 2, 98, 3]
list_num.sort()
print(list_num)
```

# Set
reference source: https://www.w3schools.com/python/python_sets.asp

```python
# create a set: unordered
fruit_set = {"apple", "banana", "cherry"}
print(fruit_set)

# access set item
for x in fruit_set:
    print(x)
    
# add item
fruit_set.add("orange")
print(fruit_set)
# add multiple item
fruit_set.update(["mango", "grapes", "peach"])
print(fruit_set)

# remove an item
fruit_set.remove("banana")
fruit_set.discard("mango")
print(fruit_set)
# not knowing which one would be popped
fruit_set.pop()
print(fruit_set)

# join 2 sets
set_alp = {"a", "b", "c"}
set_num = {1, 2, 3}
set_mix = set_alp.union(set_num)
print(set_mix)
```

# Dictionary
reference source: https://www.w3schools.com/python/python_dictionaries.asp

```python
# create a dictionary
car_dict = {
    "brand":"Ford",
    "model":"Mustang",
    "year":1964
}
print(car_dict)

# get the value of the "model" key
x = car_dict.get("model")
print(x)

# change value
car_dict["year"] = 2018
print(car_dict)

# Print all key names in the dictionary, one by one
for key in car_dict:
    print(key)
    
# Print all values in the dictionary, one by one 
for value in car_dict:
    print(car_dict[value])
for value in car_dict.values():
    print(value)
    
# Print both keys and values
for key, value in car_dict.items():
    print(key, value)
    
# add item
car_dict["color"] = "red"
print(car_dict)

# remove items
car_dict.pop("model")
print(car_dict)

# nest dictionary
child_1 = {
    "name": "Emil",
    "year":2004
}
child_2 = {
    "name":"Tobias",
    "year":2007
}
child_3 = {
    "name":"Linus",
    "year":2011
}

family_set = {
    "child1":child_1,
    "child2":child_2,
    "child3":child_3
}
print(family_set)
```
